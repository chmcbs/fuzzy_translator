### Import libraries

```python
from transformers import pipeline
```

### Test translation with Facebook NLLB

```python
# Initialise pipeline
translator = pipeline("translation", model="facebook/nllb-200-distilled-600M")
```

```python
# Test English to Korean
print(translator("Hello! Is this working?", src_lang="eng_Latn", tgt_lang="kor_Hang"))
```

```python
# Test Korean to English
print(translator("안녕하세요, 효과가 있나요?", src_lang="kor_Hang", tgt_lang="eng_Latn"))
```
